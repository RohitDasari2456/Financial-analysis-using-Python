# CAPM 


Market Weights: (the market cap of this company i) / (market cap of all company from i to j)

$ w_i = \frac{MarketCap_i}{\sum_{j}^{n}{MarketCap_j}} $

```python
tata['Cumulative'].plot(label='TATA',figsize=(16,10))
jindal['Cumulative'].plot(label='JINDAL Index')
jsws['Cumulative'].plot(label='JSWS Index')
nifty['Cumulative'].plot(label='NIFTY Index')
plt.legend()
```

# Get Daily Return

```python
tata['Daily Return'] = tata['Close'].pct_change(1)
jindal['Daily Return'] = jindal['Close'].pct_change(1)
jsws['Daily Return'] = jsws['Close'].pct_change(1)
nifty['Daily Return'] = nifty['Close'].pct_change(1)
```

```python
tata
```

```python
nifty
```

```python
plt.scatter(tata['Daily Return'],nifty['Daily Return'],alpha=0.5)
```

```python
plt.scatter(jsws['Daily Return'],nifty['Daily Return'],alpha=0.5)
```

```python
plt.scatter(jindal['Daily Return'],nifty['Daily Return'],alpha=0.5)
```

```python
tata['Daily Return'].hist(bins=100,alpha=1.0)
```

```python
nifty['Daily Return'].hist(bins=100,color='r')
```

```python
beta,alpha,r_value,p_value,std_err = stats.linregress(tata['Daily Return'].iloc[1:],nifty['Daily Return'].iloc[1:])
```
